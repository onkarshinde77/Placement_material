```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# Small vocabulary (dog, cat, apple, barks)
words = np.array(["dog", "cat", "apple", "barks"]).reshape(-1, 1)

# Create OneHotEncoder
encoder = OneHotEncoder(sparse=False)

# Fit and transform
onehot = encoder.fit_transform(words)

# Show vocabulary mapping
print("Vocabulary:", encoder.categories_)
print("\nOne-Hot Encoded Vectors:")
print(onehot)
```
